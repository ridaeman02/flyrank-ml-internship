## 1. My lane (or freestyle) and why

I select **Lane 2: Refresh / Content Opportunity Scoring**.

Many web pages lose search traffic over time (content decay). In our dataset, more than half of the pages are declining. Editors need to know which important pages to update first. This lane helps them find those pages using search and engagement data.

```python
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
print("Trend direction breakdown (percentage):")
print(df['trend_direction'].value_counts(normalize=True) * 100)
```

## 2. The question: decision, action, cost of a wrong call

- **Decision**: Which declining pages should the content team refresh or expand first?
- **Who acts**: Content editors and writers will update or rewrite the recommended pages.
- **Cost of a wrong call**: If we recommend the wrong page (false positive), the editor wastes hours updating a page that did not need it. If we miss a declining page (false negative), we lose search traffic and revenue.
- **Why ML**: We have 30,000 pages across 32 clients. There are many mixed signals like clicks, impressions, page age, and scroll rates. A simple rule is not enough to rank them all efficiently.

```python
print("Number of pages per client (showing first 5):")
print(df['client_id'].value_counts().head(5))
```

## 3. Quick look at the data (2-3 real numbers)

Here are 3 real numbers from our starter dataset:
1. **Total size**: 30,000 pages across 32 different clients.
2. **Decline rate**: 54.21% of all pages have a downward trend (`trend_direction == 'down'`).
3. **Average page age**: The average page age is 256.2 days, which shows many pages might be stale.

```python
print(f"Total pages: {len(df)}")
print(f"Unique clients: {df['client_id'].nunique()}")
print(f"Declining pages: {(df['trend_direction'] == 'down').mean() * 100:.2f}%")
print(f"Average page age: {df['content_age_days'].mean():.1f} days")
```

## 4. Careful words: what I can and can't claim

- **What we can claim**: We provide **decision-support** recommendations and **directional** rankings based on **observed** historical search visibility, position, and engagement patterns to prioritize editor review.
- **What we cannot claim**: We cannot make **causal proof** claims (e.g., "refreshing this page will cause traffic to double") because we are using observational historical data, not running randomized controlled trials. We cannot "predict Google's algorithm."
